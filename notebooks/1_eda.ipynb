```python
# Sort the data
sorted_charges = np.sort(df['charges'])

# Calculate the CDF
cdf_charges = np.arange(1, len(sorted_charges) + 1) / len(sorted_charges)

# Find the CDF value for charges less than 10k
cdf_less_than_10k = cdf_charges[sorted_charges < 10000][-1] if any(sorted_charges < 10000) else 0

print(f"The CDF for charges less than 10,000 is: {cdf_less_than_10k:.4f}")

# Optional: Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(sorted_charges, cdf_charges)
plt.axvline(x=10000, color='red', linestyle='--', label='10k')
plt.title('CDF of Charges')
plt.xlabel('Charges')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Sort the data
sorted_age = np.sort(df['age'])

# Calculate the CDF
cdf_age = np.arange(1, len(sorted_age) + 1) / len(sorted_age)

# Find the CDF value for age less than or equal to 20
cdf_less_than_equal_20 = cdf_age[sorted_age <= 20][-1] if any(sorted_age <= 20) else 0

# Probability of age greater than 20 is 1 - CDF(age <= 20)
probability_greater_than_20 = 1 - cdf_less_than_equal_20

print(f"The probability of age being greater than 20 is: {probability_greater_than_20:.4f}")

# Optional: Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(sorted_age, cdf_age)
plt.axvline(x=20, color='red', linestyle='--', label='20')
plt.title('CDF of Age')
plt.xlabel('Age')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import scipy.stats as stats

# Apply log transformation to the 'charges' column
df['log_charges'] = np.log(df['charges'])

# Plot histogram of log_charges
plt.figure(figsize=(10, 6))
sns.histplot(df['log_charges'], kde=True)
plt.title('Distribution of Log-transformed Charges')
plt.xlabel('Log(Charges)')
plt.ylabel('Frequency')
plt.show()

# Create a Q-Q plot to check for normality
plt.figure(figsize=(6, 6))
stats.probplot(df['log_charges'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Log-transformed Charges')
plt.show()
```

```python
from scipy.stats import pareto

# Fit the Pareto distribution to the charges data
# The fit function returns the shape parameter (b) and the location parameter (loc)
b, loc, scale = pareto.fit(df['charges'])

print(f"Pareto distribution parameters:")
print(f"Shape (b): {b:.4f}")
print(f"Location (loc): {loc:.4f}")
print(f"Scale (scale): {scale:.4f}")

# Plot the histogram of the charges data
plt.figure(figsize=(10, 6))
plt.hist(df['charges'], bins=50, density=True, alpha=0.6, label='Data')

# Plot the fitted Pareto distribution
# Generate x values for the fitted distribution
x = np.linspace(df['charges'].min(), df['charges'].max(), 100)
plt.plot(x, pareto.pdf(x, b, loc=loc, scale=scale), 'r-', lw=2, label='Fitted Pareto PDF')

plt.title('Fitted Pareto Distribution to Charges Data')
plt.xlabel('Charges')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import lognorm

# Filter the DataFrame for smokers and non-smokers
smokers_df = df[df['smoker'] == 'yes'].copy()
non_smokers_df = df[df['smoker'] == 'no'].copy()

# Fit log-normal distribution to charges for smokers
s_smokers, loc_smokers, scale_smokers = lognorm.fit(smokers_df['charges'], floc=0)

# Fit log-normal distribution to charges for non-smokers
s_non_smokers, loc_non_smokers, scale_non_smokers = lognorm.fit(non_smokers_df['charges'], floc=0)

# --- Plotting for Smokers ---
# Sort the charges for smokers
sorted_charges_smokers = np.sort(smokers_df['charges'])
# Calculate the empirical CDF for smokers
cdf_charges_smokers = np.arange(1, len(sorted_charges_smokers) + 1) / len(sorted_charges_smokers)

plt.figure(figsize=(10, 6))
plt.plot(sorted_charges_smokers, cdf_charges_smokers, label='Empirical CDF (Smokers)')

# Plot the theoretical log-normal CDF for smokers
x_smokers = np.linspace(smokers_df['charges'].min(), smokers_df['charges'].max(), 100)
plt.plot(x_smokers, lognorm.cdf(x_smokers, s_smokers, loc=loc_smokers, scale=scale_smokers), 'r-', lw=2, label='Fitted Log-normal CDF (Smokers)')

plt.title('CDF of Charges with Fitted Log-normal Distribution (Smokers)')
plt.xlabel('Charges')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()

# --- Plotting for Non-Smokers ---
# Sort the charges for non-smokers
sorted_charges_non_smokers = np.sort(non_smokers_df['charges'])
# Calculate the empirical CDF for non-smokers
cdf_charges_non_smokers = np.arange(1, len(sorted_charges_non_smokers) + 1) / len(sorted_charges_non_smokers)

plt.figure(figsize=(10, 6))
plt.plot(sorted_charges_non_smokers, cdf_charges_non_smokers, label='Empirical CDF (Non-Smokers)')

# Plot the theoretical log-normal CDF for non-smokers
x_non_smokers = np.linspace(non_smokers_df['charges'].min(), non_smokers_df['charges'].max(), 100)
plt.plot(x_non_smokers, lognorm.cdf(x_non_smokers, s_non_smokers, loc=loc_non_smokers, scale=scale_non_smokers), 'r-', lw=2, label='Fitted Log-normal CDF (Non-Smokers)')

plt.title('CDF of Charges with Fitted Log-normal Distribution (Non-Smokers)')
plt.xlabel('Charges')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Filter the DataFrame for non-smokers
non_smokers_df = df[df['smoker'] == 'no'].copy()

# Apply log transformation to charges for non-smokers (if not already done)
if 'log_charges' not in non_smokers_df.columns:
    non_smokers_df['log_charges'] = np.log(non_smokers_df['charges'])

# Calculate the mean and standard deviation of the log-transformed charges for non-smokers
mean_log_charges_non_smokers = non_smokers_df['log_charges'].mean()
std_log_charges_non_smokers = non_smokers_df['log_charges'].std()

print(f"Mean of log-transformed charges for non-smokers: {mean_log_charges_non_smokers:.4f}")
print(f"Standard deviation of log-transformed charges for non-smokers: {std_log_charges_non_smokers:.4f}")
```

```python
# Filter the DataFrame for smokers
smokers_df = df[df['smoker'] == 'yes'].copy()

# Apply log transformation to charges for smokers (if not already done)
if 'log_charges' not in smokers_df.columns:
    smokers_df['log_charges'] = np.log(smokers_df['charges'])

# Calculate the mean and standard deviation of the log-transformed charges for smokers
mean_log_charges_smokers = smokers_df['log_charges'].mean()
std_log_charges_smokers = smokers_df['log_charges'].std()

print(f"Mean of log-transformed charges for smokers: {mean_log_charges_smokers:.4f}")
print(f"Standard deviation of log-transformed charges for smokers: {std_log_charges_smokers:.4f}")
```

```python
# Calculate the pooled standard deviation
pooled_std = np.sqrt(((len(smokers_df) - 1) * std_log_charges_smokers**2 + (len(non_smokers_df) - 1) * std_log_charges_non_smokers**2) / (len(smokers_df) + len(non_smokers_df) - 2))

# Calculate Cohen's d
cohens_d = (mean_log_charges_smokers - mean_log_charges_non_smokers) / pooled_std

print(f"Cohen's d (effect size) for the difference in log-transformed charges between smokers and non-smokers: {cohens_d:.4f}")
```
